```python
file = open('hpepilogue.txt','r',encoding='utf-8')
text = file.read()
file.close()
wordList = text.split()
worduniq = set(wordList)

wordCount = {}

for w in worduniq:
    wordCount[w] = wordList.count(w)

del_word = ['the','a','is','of','on','that','this','and','be','to','will','we','have','in','from','our','as','not']
for w in del_word :
    if w in wordCount :
        del wordCount[w]
        
import matplotlib.pyplot as plt
from wordcloud import WordCloud

wordcloud = WordCloud(width = 1000, height = 600).generate_from_frequencies(wordCount)
plt.figure(figsize=(15,8))
plt.imshow(wordcloud)
plt.show()
```
